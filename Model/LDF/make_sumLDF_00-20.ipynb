```python
tmp = dict([('en',['10','30']),('angle',['00-20','00-20']),
            ('p',[134,31]),('h',[133,26]),('n',[129,48]),('f',[84,100])])
data = pd.DataFrame(tmp)
data
```

```python
CommonDirName = '/home/sphere/Model/Data/3RawLDF_full/'
```

```python
def calculateLDF(filen, num):
    
    filename = filen
    if(num < 10):
        filename +='00'
    elif(num < 100):
        filename +='0'
    filename += str(num)

    data = pd.read_csv(filename, header = None, index_col=None, sep='\s+')

    a = [0 for i in range(340)]
    for i in range(480):
        for j in range(480):
            dist2 = (i-239)**2 + (j-239)**2
            dist = sqrt(dist2)
            a[int (dist)] += data[i][j]

    #print(num); 
    return a
```

```python
ldf_full = []

for en in ['10', '30']:
    for par in ['p','h','n','f']:
        for angle in ['00-20']:
            
            num  = int(data[par][data['en']==en])

            filename = str(par)+'-'+en+'PeV-'+angle+'-1-'
            dirname  = str(en) + 'PeV/I/' + str(par) +'/' + angle +'/'
            name = CommonDirName+dirname+filename

            print(num, name)
            
            for k in range(num):
                ldf = calculateLDF(name, k)
                add = [en, angle, par, k] + ldf
                ldf_full.append(add)
```
