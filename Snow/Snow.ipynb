## Альбедо снега

```python
cols = ['ln_wl','alb']
albedo = pd.read_csv('./data/Grenfell.dat', header = None, names = cols,sep='\t')
print(len(albedo))
albedo['wl'] = 10 ** albedo.ln_wl 
albedo.head(2)
```

```python
plt.plot(albedo.wl, albedo.alb, alpha=1)
plt.xlim(0.3, 3)    
plt.ylim(0,1)
plt.xscale('log')
plt.xlabel('Wavelenght, $\mu$m')
plt.ylabel('Albedo')
plt.grid()
plt.show()
```

```python
# original image
display(Image('./data/Grenfell.png', width=420))
```

## Относительная спектральная мощность источника света

```python
cols = ['wl','power']
flash = pd.read_csv('./data/flash_light.dat', header = None, names = cols,sep='\t')
print(len(flash))
#flash.tail(2)
```

```python
plt.plot(flash.wl, flash.power, alpha=1)
plt.xlim(350, 800)    
plt.ylim(0,1)
plt.xlabel('Wavelenght, nm')
plt.ylabel('Relative spectral power')
plt.grid()
```

```python
# original image
display(Image('./data/flash_light.png', width=500))
```

## Спектральная чувствительность люксметра

```python
cols = ['wl','sens']
luxmeter = pd.read_csv('./data/luxmeter.dat', header = None, names = cols,sep='\t')
print(len(luxmeter))
#luxmeter.head()
```

```python
plt.plot(luxmeter.wl, luxmeter.sens, alpha=1)
plt.xlim(400, 750)    
plt.ylim(0,1)
plt.xlabel('Wavelenght, nm')
plt.ylabel('Spectral sensitivity')
plt.grid()
```
