```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.gridspec as gridspec
import numpy as np
```

```python
from datetime import timedelta
```

```python
# считать данные, заменив None на NaN
na_values = ['None', 'N/A']
data = pd.read_csv('./2018_02-03_Irkutsk.csv', na_values=na_values, index_col=False, sep='\t')
```

```python
data.head(6)
```

```python
data['SunriseTime'] = pd.to_datetime(data.Day + ' ' + data.Sunrise, dayfirst = True) - timedelta(minutes=90)
data['SunsetTime']  = pd.to_datetime(data.Day + ' ' + data.Sunset,  dayfirst = True)  + timedelta(minutes=90)
```

```python
data['MoonriseTime'] = pd.to_datetime(data.Day + ' ' + data.Moonrise, dayfirst = True)
data['MoonsetTime']  = pd.to_datetime(data.Day + ' ' + data.Moonset,  dayfirst = True) 
```

```python
def startdate(x):
    return max(x.MoonsetTime,x.SunsetTime)

data['Start'] = data.apply(startdate, axis = 1) 
```

```python
def stopdate(x):
    return min(x.SunriseTime,x.MoonriseTime)

data['Stop'] = data.apply(stopdate, axis = 1) 
```

```python
params = ['Day','SunriseTime','MoonriseTime','SunsetTime','MoonsetTime','Start','Stop']
data = data[params]
data.head(20)
```

```python
data['delta'] = data.Stop-data.Start
```

```python
data.head(64)
```

```python
for i in range(data.shape[0]-1):
    #data['Dt'][i] = 
    SR = data.SunriseTime[i]
    SS = data.SunsetTime[i]
    MR = data.MoonriseTime[i]
    MS = data.MoonsetTime[i]
    
    if MR == 'NaT':
        stop = MR
        continue
    if MS == 'NaT':
        start = MS
        continue
            
    stop  = SR    
    start = SS
    if (MS > SS) or (MS < SR):
        start = MS
    if (MR < SR) or (MR > SS):
        stop = MR
    
    print(start,'-', stop,'=', stop - start)
    #print(data.Start[i],'-', data.Stop[i+1],'=', data.Stop[i+1] - data.Start[i] )
```
