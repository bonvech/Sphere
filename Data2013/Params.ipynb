```python
def plot_for_paper(x,file, yaxislabel):
    plt.plot(datum1.minuts, datum1[x], 'm+-',  alpha=0.9, label = 'flight 1')
    plt.plot(datum2.minuts, datum2[x], 'bx--', alpha=0.9, label = 'flight 2')
    plt.plot(datum3.minuts, datum3[x], 'gs:',  alpha=0.9, label = 'flight 3')
    plt.plot(datum4.minuts, datum4[x], 'rd-.', alpha=0.9, label = 'flight 4')
    plt.plot(datum5.minuts, datum5[x], 'co-',  alpha=0.9, label = 'flight 5')
    plt.xlabel('Flight time, min')
    plt.ylabel(yaxislabel)
    plt.legend(loc=0)
    plt.grid()
    plt.savefig(file)    
```

```python
def plot_param(x, file):
    plt.figure(figsize=(8,4))
    plt.plot(datum1.minuts, datum1[x], 'm+', alpha=0.4, label = 'flight 1')
    plt.plot(datum2.minuts, datum2[x], 'bx', alpha=0.4, label = 'flight 2')
    plt.plot(datum3.minuts, datum3[x], 'gs', alpha=0.4, label = 'flight 3')
    plt.plot(datum4.minuts, datum4[x], 'rd', alpha=0.4, label = 'flight 4')
    plt.plot(datum5.minuts, datum5[x], 'co', alpha=0.4, label = 'flight 5')
    plt.xlabel('Flight time, min')
    plt.legend(loc=0)
    plt.title(x)
    plt.grid()
    if file != None:
        plt.savefig(file)
```
