```python
calibr = pd.DataFrame(columns=[str(x) for x in range(109)])
events = []

num = 0 
for event in range(10675, 14487):
    filename = '../calibr/' + str(event) +'.cal'
    try:
        one = pd.read_csv(filename, header = None, index_col=False, sep='\s+')
    except:
        #print(event, "No file", filename)
        continue
    calibr.loc[num] = one[13].values[:109]
    events.append(event)
    #calibr.loc[num,"event"] = int(event)
    num += 1
calibr["event"] = events

columns = ["event"]
columns += [str(x) for x in range(109)]
calibr = calibr[columns]
print(calibr.shape)
```

```python
calibr
```

```python
calibr.to_csv("calibr.csv")
```

```python
''' Нарисовать калибровки'''
for pmt in range(31,32):
    data = calibr[str(pmt)]
    #data = data[abs(data)<1000]
    plt.plot(data)
```

```python
plt.rcParams['figure.figsize'] = (12,4)

for pmt in range(109): 
    fig, ax = plt.subplots()
    data = calibr[str(pmt)]
    #data = data[abs(data)<1000]
    plt.plot(data)
    ax.set_title("PMT" + str(pmt+1))
    plt.grid()
```
